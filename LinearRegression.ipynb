# Linear Regression Project

This notebook demonstrates a junior-level end-to-end machine learning workflow for a linear regression problem.

It includes:
- preparing a regression dataset
- performing exploratory data analysis (EDA)
- training a linear regression model
- evaluating accuracy and error metrics
- visualizing the relationship between features and the target

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

plt.style.use('seaborn-v0_8')
sns.set_theme(style='whitegrid')

np.random.seed(42)

# Create a synthetic regression dataset with a strong linear relationship
n_samples = 200
feature_x = np.random.uniform(10, 100, n_samples)
noise = np.random.normal(0, 8, n_samples)
target_y = 3.5 * feature_x + 20 + noise

df = pd.DataFrame({'area': feature_x, 'price': target_y})

print(df.head())
print('\nDataset shape:', df.shape)
```

```python
# Basic statistics and data quality checks
print(df.describe())
print('\nMissing values:\n', df.isnull().sum())

# Visualize the distribution of features and target
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df['area'], kde=True, ax=axes[0])
axes[0].set_title('Distribution of Area')
axes[0].set_xlabel('Area')
axes[0].set_ylabel('Frequency')

sns.histplot(df['price'], kde=True, ax=axes[1])
axes[1].set_title('Distribution of Price')
axes[1].set_xlabel('Price')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Scatter plot to understand linear relationship
plt.figure(figsize=(7, 4))
sns.scatterplot(data=df, x='area', y='price')
plt.title('Area vs Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# Prepare the data for model training
X = df[['area']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print('Training set shape:', X_train.shape)
print('Test set shape:', X_test.shape)
```

```python
# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, predictions)

print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_[0])
print('MAE:', round(mae, 2))
print('MSE:', round(mse, 2))
print('RMSE:', round(rmse, 2))
print('R2 Score:', round(r2, 4))
```

```python
# Visualize model fit
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='area', y='price', alpha=0.7, label='Observed')
plt.plot(sorted(X_test['area']), model.predict(pd.DataFrame({'area': sorted(X_test['area'])})), color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()

# Residual plot
residuals = y_test - pd.Series(predictions, index=y_test.index)
plt.figure(figsize=(8, 4))
sns.scatterplot(x=y_test.index, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Sample Index')
plt.ylabel('Residual')
plt.show()
```
